# 2. The Data Analysis Approach

We will use a three-step approach to data analysis with an AI assistant:
1. **Decomposition**: Break down the high-level task into a sequence of smaller subtasks. 
    - Each subtask should take as inputs either raw data files (.csv files) or outputs from previous subtasks.
    - The output of each subtask except the last will be used as input to a later subtask.
    - Give each subtask a descriptive name.
2. **Create Test Data**: Create small test versions of each of the .csv files that contain just enough data to be able to tests the correctness of the data analysis.
3. **Write Down Correct Results for Test Data**
    - Assume you have already implemented code that implements the subtasks (but don't actually implement it yet).
    - Imagine running the code for sequence of subtasks, passing in the test data files as inputs.
    - Write down what the correct output of each subtask should be, assuming it gets as input the test data files or the correct outputs of previous subtasks.
4. **Implement Subtasks**: For each subtask, write code that implements it.
5. **Run on Test Data**: Run the code on the test data files and check that the outputs match the correct outputs you wrote down.
6. **Run on Full Data**: Once the code works on the test data, run it on the full data files to get the final results.

**You will use your AI assistant to help you with all of these steps! But feed it one of these tasks at a time, so if it gets something wrong, it will be easier to fix.**
